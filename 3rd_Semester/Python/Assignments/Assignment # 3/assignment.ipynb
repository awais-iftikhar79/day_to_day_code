```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Section 1

```python
days = np.array(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
screen_time = np.array([4.85,4.78,4.88,3.53,7.51,6.68,7.26])
whatsapp = np.array([0.8,0.71,0.51,0.58,2.25,3.56,0.61])  # hours + mints/60
insta = np.array([2.48,1.71,3.15,1.65,3.36,2.16,5.88])
facebook = np.array([1.03,1.53,0.98,1.11,1.4,0.36,0.68])
notifiation = np.array([458,440,398,453,773,745,397])
apps_open = np.array([119,120,94,117,158,186,85])
```

```python
data = {
    'Days':days ,
    'TotalScreenTime':screen_time ,
    'WhatsApp':whatsapp ,
    'Instagram': insta,
    'FaceBook': facebook,
    'Notifications':notifiation ,           
    'AppsOpen':apps_open ,                  
}

df = pd.DataFrame(data)
print(df)
```

```python
#checking any null in data set
df.isnull().sum()
```

As above ouput clearly shows that data set is already properly cleaned (You can also see this in screeshots as proof)

Section 2

Line Plot 

```python
plt.figure(figsize=(8,6))  # i add this because on my axis Day's Name are too much closer thats why i give dimsesion of of convas myself

plt.plot(df['Days'],df['TotalScreenTime'],marker='o',color='lightgreen')
plt.xlabel('Days',fontsize=12,color='blue')
plt.ylabel('Screen Time',fontsize=12,color='blue')
plt.title('Daily Screen Time',fontsize=16,color='blue')
plt.grid(True,alpha=0.4)
plt.show()
```

My highest usage occurred on Friday with 7.51 hours
The lowest screen time was on Thursday with 3.53 hour

Bar Chart

```python
index_day = np.where(df['Days']=='Friday')[0][0]

whats_usage=whatsapp[index_day]
insta_usage=insta[index_day]
face_usage=facebook[index_day]

apps = ['WhatsApp','Instagram','FaceBook']
usage = [whats_usage,insta_usage,face_usage]

plt.bar(apps,usage)
plt.xlabel('Apps',fontsize=12)
plt.ylabel('Usage in Hours',fontsize=12)
plt.title('App usage on Friday',fontsize=16)
plt.show()
```

WhatsApp  usage on Friday 2.25 hours

Instagram usage  on Friday 3.36 hours

Facebook  usage on Friday 1.40 hours

Histogram

```python
plt.hist(df['TotalScreenTime'],edgecolor='black')
plt.xlabel('Screen Time (Hours)',fontsize=12)
plt.ylabel('Frequency (Number of Days)',fontsize=12)
plt.title('Distribution of Screen Time Across Days',fontsize=16)
plt.grid(axis='y',alpha=0.4,linestyle='--')
plt.show()
```

The histogram shows that most days have screen time around 4.5 to 5 hours, with fewer days having very low or high screen time

Scatter Plot

```python
plt.scatter(df['TotalScreenTime'],df['Notifications'],color='orange')
plt.xlabel('Screen Time (hours)',fontsize=12)
plt.ylabel('Number of Notifications',fontsize=12)
plt.title('Screen Time vs Notifications',fontsize=16)
plt.grid(True)
plt.show()
```

I noticed that on days with high notifications, my screen time was also higher

Pie Chart

```python
total_whats = df['WhatsApp'].sum()
total_insta = df['Instagram'].sum()
total_face = df['FaceBook'].sum()

apps = ['WhatsApp','Instagram','FaceBook']
weekly_usage = [total_whats,total_insta,total_face]

plt.pie(weekly_usage,labels=apps,autopct='%1.1f%%',startangle=90,shadow=True,explode=(0.04,0.06,0.04))
plt.title("Weekly App Usage Distribution")
plt.legend(loc='lower left',fontsize=7)
plt.show()
```

Instagram has the highest usage 55.9% of the total weekly app usage
