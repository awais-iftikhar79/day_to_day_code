Data Cleaning

```python
df = pd.DataFrame({
    'Name': ['Awais', 'Ali', 'Hammad', None],
    'Age': [22, None, 21, 20],
    'City': ['Topi', 'Lahore', None, 'Karachi']
})
```

```python
print("Sum of Null Values in Each Coloumn :\n",df.isnull().sum())
```

```python

df['Name'].fillna('Unknown',inplace=True)
df['Age'].fillna(0,inplace=True)
df['City'].fillna('No City',inplace=True)

df
```

```python
greater_20 = df[df['Age'] > 20]
print(greater_20)
```

```python
print(df.loc[0:2,['Name','City']])
```

```python
print(df.iloc[1:3:,1])
```

Sorting , Merging

```python
df = pd.DataFrame({
    'Name': ['Awais', 'Ali', 'Awais', 'Hammad', 'Ahmed', 'Awais'],
    'Age': [22, 21, 22, 21, 23, 24],
    'Marks': [90, 85, 90, 88, 75, 95],
    'City': ['Topi', 'Lahore', 'Topi', 'Karachi', 'Lahore', 'Topi']
})

print(df)
```

```python
#sorting by age 
df = df.sort_values('Age',ascending=False)
print(df)
```

```python
#remove duplicates
df = df.drop_duplicates(subset=['Name'],keep='last')
print(df)
```
