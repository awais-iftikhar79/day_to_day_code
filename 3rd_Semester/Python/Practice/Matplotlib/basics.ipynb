```python
import matplotlib.pyplot as plt
```

```python
#Simple Plot
x=[1,2,3,4,5]
y=[5,10,8,12,20]

plt.plot(x,y)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales Over Time')
plt.grid(True)
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
temperature = [30,32,31,29,28,27,30]

plt.plot(days,temperature)
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Temperature/Days')
plt.grid(True)
plt.show()

days = [1,2,3,4,5]
sales = [100,120,90,150,170]

plt.figure(figsize=(6,5))

plt.plot(days,sales)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Sales/Days')
plt.grid(True)
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25, 35]

plt.plot(x,y,marker='o',color='red',linestyle='--')
plt.show()
```

Bar

```python
Products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor']
Sales = [80, 120, 60, 100]

plt.bar(Products,Sales,color='lightgreen',label='Product Count')

plt.xlabel('Products',fontsize=12)
plt.ylabel('Sales',fontsize=12)
plt.title('Product Count',fontsize=16)

plt.grid(axis='y',linestyle='--',alpha=1)
plt.legend()
plt.show()
```

Pie

```python
Categories = ['Electronics', 'Clothes', 'Grocery', 'Furniture']
Sales = [400, 200, 300, 100]

plt.pie(Sales,labels=Categories,autopct='%1.f1%%',startangle=90)
plt.title('Pie Chart')
plt.show()
```
