University Management System

```python
class Person:
    def __init__(self,name,id):
        self._name = name
        self._id = id
        
    def display_info(self):
        print(f'Name : {self._name}\nID : {self._id}')

```

Derived student class

```python
class Student(Person):
    def __init__(self, name, id,cgpa):
        super().__init__(name, id)
        self.cgpa = cgpa
        self._courses = []
        
    @property
    def cgpa(self):
        return self._cgpa
    
    @cgpa.setter
    def cgpa(self,value):
        if(value<0 or value>4):
            raise ValueError('cgpa cannot be negative')
        self._cgpa = value
        
    def add_course(self,course):
        if course not in self._courses:
            self._courses.append(course)
            print(f'{course} added successfully !')
        else:
            print(f'{course} already registered')
            
    def display_info(self):
        super().display_info()
        print(f'CGPA : {self._cgpa}')
        if self._courses:
            for course in self._courses:
                print(f'-{course}')
        else:
            print('Not registered')
            
    
```

Derived Teacher Class

```python
class Teacher(Person):
    def __init__(self, name, id,salary):
        super().__init__(name, id)
        self.salary = salary
        
    @property
    def salary(self):
        return self._salary
    
    @salary.setter
    def salary(self,value):
        if(value<0 ):
            raise ValueError('Salray cannot be negative')
        self._salary = value
        
    def display_info(self):
        super().display_info()
        print(f'Salary : {self._salary}')
```
